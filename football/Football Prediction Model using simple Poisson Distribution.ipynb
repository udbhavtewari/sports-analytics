# Taking Chelsea (Home) vs Manchester United (Away)

```python
# Premier League Calculations
htg = df['FTHG'].sum() # Total Home Team Goals
atg = df['FTAG'].sum() # Total Away Team Goals
tot_matches = len(df.index) # Total Matches
goals_per_matches_pl_home = htg/tot_matches
goals_per_matches_pl_away = atg/tot_matches
```

```python
# Chelsea Calculations
chl = df.query("HomeTeam=='Chelsea'") # Selecting Chelsea as home team

chl_home_goals_scored = chl['FTHG'].sum() # Home goals scored by Chelsea

chl_home_goals_conceded = chl['FTAG'].sum() # Home goals conceded by Chelsea

goals_per_matches_chl_home_scored = chl_home_goals_scored/len(chl) # Goals per match for Chelsea at home scored

goals_per_matches_chl_home_conceded = chl_home_goals_conceded/len(chl) # Goals per match for Chelsea at home conceded


```

```python
# Man Utd Calculations
manu = df.query("AwayTeam=='Man United'") # Selecting Man Utd as away team

manu_away_goals_scored = manu['FTAG'].sum() # Home goals scored by Man Utd

manu_away_goals_conceded = manu['FTHG'].sum() # Home goals conceded by Man Utd

goals_per_matches_manu_away_scored = manu_away_goals_scored/len(manu) # Goals per match for Man Utd at away scored

goals_per_matches_manu_away_conceded = manu_away_goals_conceded/len(manu) # Goals per match for Man Utd at away conceded
```

# Attack Strength 
*Team's avg number of goals divided by the league's avg number of goals*

```python
chl_att = goals_per_matches_chl_home_scored/goals_per_matches_pl_home
man_att = goals_per_matches_manu_away_scored/goals_per_matches_pl_away
```

# Defence Strength
*Team's avg number of goals conceded divided by the league's avg number of goals conceded*

```python
chl_def = goals_per_matches_chl_home_conceded/goals_per_matches_pl_away
man_def = goals_per_matches_manu_away_conceded/goals_per_matches_pl_home
```

# Goal Expectency
*To determine likely goals for each team*
* For Chelsea -> Chelsea's attack strength x Man Utd's defence strength * league's avg number of goals
* For Man Utd -> Man Utd's attack strength x Chelsea's defence strength * league's avg number of goals


```python
chl_exp = chl_att * man_def * goals_per_matches_pl_home # Chelsea's goal expectancy 

man_exp = man_att * chl_def * goals_per_matches_pl_away # Man Utd's goal expectancy 

print ("Chelsea's goal expectancy : " + str(chl_exp))
print ("Man Utd's goal expectancy : " + str(man_exp))
```

# Poisson Distribution
*Since no match ends with the above goal expectancy, Poisson Distribution allows us to use these figure to distribute 100% of probability across a range of goal outcomes for each side*

```python
from scipy.stats import poisson
```

### Probability of Chelsea scoring 0 upto 5 goals against Man Utd

```python
chl_prob = []
q = 0
while q < 6 :
    b = poisson.pmf(q,mu=1.0416666666666667)
    q = q + 1
    chl_prob.append(b)

    
```

### Probability of Man Utd scoring 0 upto 5 goals against Chelsea

```python
manu_prob = []
k = 0
while k < 6 :
    a = poisson.pmf(k,mu=0.95610204550678)
    k = k + 1
    manu_prob.append(a)
    
```

# Predicting the match outcomes based on these probablities
*To get each possible score, multiply the probablity of each possible score of each team by the probability of each possible score by the other team*

```python
goals = [0,1,2,3,4,5]
list_of_tuples = list(zip(goals,chl_prob, manu_prob)) 
```

```python
list_of_tuples
```

```python
prob_table = pd.DataFrame(list_of_tuples,
                  columns = ['Goals','Chelsea Probability', 'Man Utd Probability']) 
```

```python
# Predicting the match outcome probability 
prob_table['Match Probability'] = prob_table['Chelsea Probability'] * prob_table['Man Utd Probability']
```
